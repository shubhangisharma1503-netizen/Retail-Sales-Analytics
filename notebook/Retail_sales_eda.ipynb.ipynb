```python
## feature engineering means creating new columns from existing data to make analysis easier and more meaningful...
## for example, we can create a new column called "Order Month" from the "Order Date" column to analyze sales by month.
```

```python
df["Year"]=df["Order Date"].dt.year
df["Month"]=df["Order Date"].dt.month_name()
df["Quarter"]=df["Order Date"].dt.quarter
df["Day"]=df["Order Date"].dt.day_name()
#dt is datetime accessor in pandas, it allows us to access the year, month, day, etc. of a datetime column.
```

```python
df[["Order Date","Year","Month","Quarter","Day"]].head()
```

```python
df["Shipping Duration"] = (df["Ship Date"] - df["Order Date"]).dt.days
```

```python
df["Profit Margin"] = (df["Profit"] / df["Sales"]) * 100
```
